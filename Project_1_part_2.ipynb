# importing numpy and panda

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data_preprocessed
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
data_preprocessed
```

```python
target = np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
data_preprocessed['Excessive Absenteeism'] = target
data_preprocessed
```

```python
target.sum()/target.shape[0]
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 'Day of the Week', 'Distance to Work',
        'Daily Work Load Average'],axis =1)
data_with_targets
```

```python
data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
unscaled_inputs
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

# split the data in train and test then shuffle

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs,target)
```

```python
x_train,x_test,y_train,y_test  = train_test_split(scaled_inputs, target, train_size = 0.8)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

# Logistic Regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

## Manually checking accuracy

```python
model_outputs = reg.predict(x_train)
model_outputs
```

```python
y_train
```

```python
model_outputs==y_train
```

```python
np.sum(model_outputs==y_train)
```

```python
model_outputs.shape[0]
```

```python
np.sum(model_outputs==y_train)/model_outputs.shape[0]
```

## Finding intercepts and coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns = ['Features'],data = feature_name)
summary_table['Coefficient' ] = np.transpose(reg.coef_)
summary_table.index = summary_table.index+1
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

## Interpreting the coefficients 

```python
summary_table['Odds_Ratio']= np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds_Ratio',ascending = False)
summary_table
```

# testing the model

```python
reg.score(x_test,y_test)
```

```python
predicted_proba = reg.predict_proba(x_test)
predicted_proba
```
